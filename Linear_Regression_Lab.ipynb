```python
features=['curbweight','enginesize','horsepower','doornumber_two','drivewheel_rwd']
features2=['fuelsystem_idi','compressionratio','fueltype_gas','cylindernumber_two','enginetype_rotor']
y=carsData2['price']
import numpy as np
# Fit your model

model = LinearRegression().fit(carsData2[features], y)


# Get R^2
r2 = model.score(carsData2[features], y)


# n = number of rows, p = number of predictors
n = carsData2[features].shape[0]
p = carsData2[features].shape[1]


# Adjusted R^2
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)


print("R^2:", r2)
print("Adjusted R^2:", adj_r2)
```
