When comparing the sample size to the actual data, the mean of the sample size looks to be between 0 - 0.1 while the mean of the acutal dataset is about 0.7. Also, I believe there is some bias to the sample data because we sorted by 'duration_ms'.  If we were to randomly select 50 sample from the dataset compared to using the top50 after sorting it by 'duration_ms', we might get a better estimate of the actual mean.

```python
sample_size=list(range(10,501))
rel_errors=[]
for i in sample_size:
  sample_mean=df_spot.duration_ms.sample(n=i, random_state=2026).mean()
  relative_error= abs(sample_mean-true_duration_mean)/true_duration_mean
  rel_errors.append(relative_error)
```

##22

```python
n=100
die_1=np.random.choice((1,2,2,3,3,4,7,9),size=n, replace=True)
die_2=np.random.choice((1,2,3,4,5,6),size=n, replace=True)
die_samp_100=(die_1+die_2)/2
```

```python
die_samp_100
```

```python
point_estimate=die_samp_100.mean()
SE_2=np.std(die_samp_100, ddof=1)
SE_2=SE_2/np.sqrt(n)
SE_2
```

```python
from scipy.stats import norm
die_lower=norm.ppf(.025,point_estimate, SE_2)
die_upper=norm.ppf(.975,point_estimate, SE_2)
print((die_lower, die_upper))
```

```python
point_estimate
```

```python
SE_2
```

```python
moe= (die_upper-die_lower)/2
moe
```
